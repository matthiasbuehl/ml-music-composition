```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.model_selection import train_test_split

import numpy as np
import matplotlib.pyplot as plt
import pprint as pp

#pp = pprint.PrettyPrinter(indent=2)
```

```python

data = [[[(i+j) / 100] for i in range(5)] for j in range(100)]
data = np.array(data, dtype=float)

labels = [(i+5) / 100 for i in range(100)]
labels = np.array(labels, dtype=float)

print(data.shape)
print(labels.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=4)
```

```python
model = Sequential()
model.add(LSTM((1), batch_input_shape=(None, 5, 1), activation='relu', return_sequences=False))
#model.add(LSTM((1), activation='relu', return_sequences=False))
#model.add(Dense(1, activation='relu'))
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
print(model.input_shape)
print(model.output_shape)
print(model.summary())
```

```python
history = model.fit(x_train, y_train, epochs=1000, validation_data=(x_test, y_test))
```

```python
results= model.predict(x_test)
print(results)
print(y_test)
```

```python
plt.scatter(range(20), results, c='r')
plt.scatter(range(20), y_test, c='g')
```
